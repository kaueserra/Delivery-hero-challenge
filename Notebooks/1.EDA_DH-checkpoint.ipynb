```python
ds1 = pd.read_csv("C:/Users/Kaue/Reicarnation-Blues-Book/data/machine_learning_challenge_order_data.csv",low_memory=False)  

ds2 = pd.read_csv("C:/Users/Kaue/Reicarnation-Blues-Book/data/machine_learning_challenge_labeled_data.csv",low_memory=False)  

ds = pd.merge(ds1,ds2,how='left',on='customer_id')   

ds['voucher_bol'] = np.where(ds['voucher_amount']>0,1,0) 

ds['delivery_free'] = np.where(ds['delivery_fee']>0,1,0)


#basic data transformation

ds['order_date'] = pd.to_datetime(ds['order_date'],format="%Y-%m-%d")   

ds.head()
```

# EDA time 
    - Even thought this is a classification problem, there can be time related features, 
    therefore it be plotted some time series as well

```python
ds['date_agg'] = ds['order_date'].values.astype('datetime64[M]') 


ds_agg_dt = pd.DataFrame(ds.groupby('date_agg').agg({  
                         'customer_id' : 'count', 
                          'is_failed':'mean', 
                         'voucher_amount' : 'sum' , 
                         'delivery_fee' : 'mean',
                          'amount_paid' : 'mean',
                         'restaurant_id' : lambda x:x.nunique(),
                         'city_id' : lambda x:x.nunique(), })) 


var_set = ['total_transactions','avg_failed','total_voucher','avg_delivery_fee','avg_amount','restaurants','cities']


ds_agg_dt = ds_agg_dt.set_axis(var_set, axis=1, inplace=False)

print(ds_agg_dt.shape)  

print('\n' + '¹'*40)

# Let's analyse at least 100 transactions per month, otherwise the other metrics will be a bit off

ds_agg_dt = ds_agg_dt[ds_agg_dt.total_transactions>100]  

print('Months with enough data are {}'.format(ds_agg_dt.shape[0])) 

print('\n' + '²'*40)

ds_agg_dt.head()
```

```python
fig, ax = plt.subplots(7) 
fig.set_size_inches(14, 40)   


aux = 0
for i in var_set: 
    
        ax[aux].plot(ds_agg_dt[i],color=random_color()) 
        #ax[aux].xaxis.set_major_locator(mdates.MonthLocator())   
        ax[aux].fmt_xdata = mdates.DateFormatter('%Y-%m-%d')
        ax[aux].set_title('{}'.format(i) )
        aux += 1





    
```

```python
vars_dist = ['customer_id','restaurant_id','city_id']

fig, axs = plt.subplots(ncols=2,nrows=len(vars_dist)) 
fig.set_size_inches(14, 25)   


i = 0 

for var in vars_dist:

    for x in (0,1):
        # Subset to the airline
        subset = ds[ds['is_returning_customer'] == x]

        # Draw the density plot
        sns.distplot(subset[var].value_counts(),hist=False, kde=True,
                     kde_kws = {'linewidth': 1},
                     label = x, ax=axs[i][0])   
        axs[i][0].set_title('Distribution of the feature {}'.format(var))


        # Draw the density plot
        sns.distplot(subset[var].value_counts(),hist=False, kde=True,
                     kde_kws = {'linewidth': 1},
                     label = x, ax=axs[i][1]) 

        axs[i][1].set_xlim(0,50*(i*2+1)) 
        axs[i][1].set_title('Zoom for the features {} for below 50'.format(var)) 

    i+=1
```

```python
col_cat = ['order_hour','is_failed','payment_id', 'platform_id',
           'transmission_id','voucher_bol','delivery_free'] 

fig, axs = plt.subplots(ncols=2,nrows=len(col_cat))
fig.set_size_inches(15, 25)  

j=0 

aux =0 

for var1 in col_cat:  
    
    sns.catplot(x=var1, kind="count", data=ds, ax=axs[j][0]) 
    axs[j][0].set_title('Frequency of the categories of {}'.format(var1)) 
    plt.close()
    
    
    sns.catplot(x=var1, y='is_returning_customer', kind="bar", data=ds, ax=axs[j][1])  
    axs[j][1].set_title('Proportion of each category to returning customer')
    plt.close()

    j+=1 

    
plt.subplots_adjust(hspace=0.6)
plt.show()


```

## EDA on the customer level 

Now that Analysing the transactions we could notice that, there are different distribution of the number of transactions of returning clients against the ones that do no go back. 

So the first step is to aggregate the variables on customer level, and with this a lot of new features will be created to represent the other dimensions of one feature

```python
def groupped_last_months(months):
    ds = ds.sort_values(by=['order_date','customer_id'])

    last_month_ds = ds[ds.order_date > (today- dateutil.relativedelta.relativedelta(months=6)) ] 


    cols_name = ['orders_m1','last_m1','failed_sum_m1','delivery_fee_mean_m1',
              'voucher_amount_sum_m1','last_rest_m1','last_city_m1','last_payment_m1',
              'last_platform_m1','transmission_id_m1','voucher_bol_m1','delivery_free_m1']

    last_month_ds_agg =  pd.DataFrame(ds.groupby('customer_id').agg({  
                             'customer_id' : 'count',  
                              'order_date': 'first', 
                              'is_failed':'sum', 
                             'voucher_amount' : 'sum' , 
                             'delivery_fee' : 'mean',
                              'amount_paid' : 'mean',
                             'restaurant_id' : 'first',
                             'city_id' : 'first' ,
                             'payment_id': 'first',
                            'platform_id':'first',
                           'transmission_id':'first',  
                            'voucher_bol': 'sum', 
                            'delivery_free': 'sum', }),columns= cols_name)   
    
    return last_month_ds_agg



```
