```python
ds_test =  pd.DataFrame(ds.groupby('customer_id').agg({ 
                        'order_date':'max', 
                        'is_returning_customer':'max'}))


today = datetime.strptime('2017-02-28', '%Y-%m-%d') 

ds_test['diff'] = (today - ds_test.order_date).dt.days 


cut_labels = ['<1 m','1 - 3m','3m - 6m', '6 - 12m','+12']
cut_bins =[0,31, 90, 180, 365,2000]

ds_test['diff_cat'] = pd.cut(ds_test['diff'],
                            bins=cut_bins,
                            labels=cut_labels) 

fig, axs = plt.subplots(nrows=2)   
fig.set_size_inches(10,8) 

sns.countplot(x="diff_cat", data=ds_test,ax=axs[0])    
axs[0].set_title('Frequency and % of returning customer for the last transaction*')

sns.barplot(x="diff_cat", y="is_returning_customer",ax=axs[1], data=ds_test, estimator=lambda x: sum(x) / len(x) * 100)
axs[1].set_ylabel('Percent')
axs[1].set_xlabel("Frequecy of the last transaction per customer") 

# We can see here that customers with recent transactions have the tendecy to return,  
# And most of the costumer made the last transaction more than a year ago
```

```python
ds = ds.sort_values(by=['order_date','customer_id'])

def groupped_last_months(mon):
    
    
    last_month_ds = ds[ds.order_date > (today- dateutil.relativedelta.relativedelta(months=mon)) ] 

    cols_name = ['orders_m','last_m','failed_sum_m','voucher_amount_sum_m','delivery_fee_mean_m',
              'amount_paid_m','last_rest_m','last_city_m','last_payment_m',
              'last_platform_m','transmission_id_m','voucher_bol_m','delivery_free_m']
 

    cols_name = [sub + str(mon) for sub in cols_name]  
    
    
    last_month_ds_agg =  pd.DataFrame(ds.groupby('customer_id').agg({  
                             'customer_id' : 'count',  
                              'order_date': 'first', 
                              'is_failed':'sum', 
                             'voucher_amount' : 'sum' , 
                             'delivery_fee' : 'mean',
                              'amount_paid' : 'mean',
                             'restaurant_id' : 'first',
                             'city_id' : 'first' ,
                             'payment_id': 'first',
                            'platform_id':'first',
                           'transmission_id':'first',  
                            'voucher_bol': 'sum', 
                            'delivery_free': 'sum', }))    
    
    
    last_month_ds_agg.columns= cols_name
    
    return last_month_ds_agg



```
